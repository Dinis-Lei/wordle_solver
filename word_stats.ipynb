```python
lst = []
with open('stats0.txt') as f:
    for line in f:
        lst.append(int(line))

print(sum(lst)/len(lst))
print(max(lst))
print(min(lst))
print(sum([x < 6 for x in lst]))
print(sum([x >= 6 for x in lst]))
```
