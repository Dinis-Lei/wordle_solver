```python
def get_stats(stat):
    lst = []
    with open(stat) as f:
        for line in f:
            lst.append(int(line))

    print(sum(lst)/len(lst))
    print(max(lst))
    print(min(lst))
    print(sum([x < 6 for x in lst]))
    print(sum([x >= 6 for x in lst]))
```

```python
get_stats('stats0.txt')
print("-"*10)
get_stats('stats1.txt')
print("-"*10)
get_stats('stats2.txt')
print("-"*10)
get_stats('stats3.txt')
print("-"*10)
get_stats('stats4.txt')
```
